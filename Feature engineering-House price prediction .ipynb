### Feature engneering


We will be performing all the below steps in Feature Engineering

Missing values<br>
Temporal variables<br>
Categorical variables: remove rare labels<br>
Standarise the values of the variables to the same range

- 1. Handling missing values
     - numerical features
     - categorical features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('train_AHPP.csv')
pd.set_option('display.max_columns',None)
dataset.head()
```

```python
#categorical features with nan
categorical_with_nan =[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtype=='O']
for feature in categorical_with_nan:
    print(feature,': ',dataset[feature].isnull().mean()*100," %")
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,categorical_with_nan):
    data=dataset.copy()
    data[categorical_with_nan]= data[categorical_with_nan].fillna('Missing')
    return data

dataset =replace_cat_feature(dataset,categorical_with_nan)
dataset.head()
```

```python
dataset[categorical_with_nan].isnull().sum()
```

```python
#lest just handle missing values of numerical feature

numerical_feature_with_nan =[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
for feature in numerical_feature_with_nan:
    print(feature," :",dataset[feature].isnull().mean()*100,"%")
```

```python
for feature in numerical_feature_with_nan:
    median =dataset[feature].median()
    
    #capturing the missing values
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median,inplace=True)
```

```python
for feature in numerical_feature_with_nan:
    print(feature," :",dataset[feature].isnull().mean()*100,"%")
```

```python
dataset.head(10)
```

### Temporal variable

```python
year_feature = ['YearBuilt','YearRemodAdd','GarageYrBlt']
```

```python
year_feature
```

```python
for feature in year_feature:
    dataset[feature] = dataset['YrSold']-dataset[feature]
    
        
        
```

```python
dataset[year_feature]
```

```python
dataset.head()
```

### Numerical Variables
Since the numerical variables are skewed we will perform log normal distribution

```python
numerical_variable= ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in numerical_variable:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

### Handling Rare Categorical Feature
We will remove categorical variables that are present less than 1% of the observations

```python
categorical_var = [feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_var
```

```python
#example
l=dataset.groupby('Street')['SalePrice'].count()/len(dataset)
l[l>0.5].index
```

```python
for feature in categorical_var:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(10)
```

```python
#example
label =dataset.groupby(['Utilities'])['SalePrice'].mean().sort_values().index
{k:i for i,k in enumerate(label,0)}
```

```python
for feature in categorical_var:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

### Feature Scaling

```python
feature_scale =[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),
                pd.DataFrame(scaler.transform(dataset[feature_scale]),columns=feature_scale)],axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```

### Now we will perform the same steps for test dataset
